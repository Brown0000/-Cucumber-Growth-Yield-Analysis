## 4. Bar Chart Plotting Function

```python
def plot_growth_over_weeks(title, data, weeks):
    treatments = list(data.keys())
    width = 0.2
    x = np.arange(len(weeks))

    for i, treatment in enumerate(treatments):
        plt.bar(x + (i - 1.5) * width, data[treatment], width, label=treatment)

    plt.xlabel("Weeks")
    plt.ylabel(title)
    plt.title(title)
    plt.xticks(x, weeks)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

## 5. Generate Growth Charts

```python
# Plant Height
plot_growth_over_weeks("Plant Height (March to July 2024)", plant_height_march_july, weeks)
plot_growth_over_weeks("Plant Height (August to October 2024)", plant_height_aug_oct, weeks)

# Number of Leaves
plot_growth_over_weeks("Number of Leaves (March to July 2024)", leaves_march_july, weeks)
plot_growth_over_weeks("Number of Leaves (August to October 2024)", leaves_aug_oct, weeks)

# Stem Diameter
plot_growth_over_weeks("Stem Diameter (March to July 2024)", stem_diameter_march_july, weeks)
plot_growth_over_weeks("Stem Diameter (August to October 2024)", stem_diameter_aug_oct, weeks)
```
